```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Kyleinfo=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/refs/heads/main/ComplaintsTable.csv")
Kyleinfo.head()
```

```python
total_complaints=Kyleinfo.groupby("ProductCategory")["ComplaintCount"].sum()
total_complaints
```

```python
sorted_complaints=total_complaints.sort_values(ascending=False)
sorted_complaints
```

```python
cumulative_sum=sorted_complaints.cumsum()
cumulative_sum
```

```python
cumulative_percentage=round(cumulative_sum/sorted_complaints.sum()*100,2)
cumulative_percentage
```

```python
fig, ax1=plt.subplots()
sorted_complaints.plot(kind='bar',color='C0',ax=ax1)
ax1.set_ylabel("ComplaintCount")
ax2=ax1.twinx()
cumulative_percentage.plot(kind='line',color='C1',ax=ax2,marker='D')
ax2.set_ylabel('Cumulative Percentage %')
plt.title("Pareto Analysis for the ComplaintsCount")
```

```python
Top4=Kyleinfo.query('ProductCategory in ["Kitchenware","Bedding","Jewelry","Books"]')
Top4
```

```python
sorted_complaints1=Top4.groupby("ComplaintReason")["ComplaintCount"].sum().sort_values(ascending=False)
sorted_complaints1
```

```python
cumulative_complaint_Percentage=round(sorted_complaints1.cumsum()/sorted_complaints1.sum()*100,2)
cumulative_complaint_Percentage
```

```python
fig, ax1=plt.subplots()
sorted_complaints1.plot(kind='bar',color='C0',ax=ax1)
ax1.set_ylabel("ComplaintCount")
ax2=ax1.twinx()
cumulative_complaint_Percentage.plot(kind='line',color='C1',ax=ax2,marker='D')
ax2.set_ylabel('Cumulative Percentage %')
plt.title("Paroto analysis of the ComplaintReason associated with returns")
```
